```python
import pandas as pd
import requests
import csv
from urllib.parse import urlparse
from tqdm import tqdm

df = pd.read_csv('/kaggle/input/trends-batch4/full_merged_data_trivial.csv')
package_names = df['package_name'].tolist()

package_names = package_names[6400:]
print(len(package_names))
data_rows = []

for package_name in tqdm(package_names):
    try:
        url = f'https://registry.npmjs.org/{package_name}'

        # Fetching the full metadata response
        response = requests.get(url)
        response.raise_for_status()

        package_data = response.json()
        
        # Fetching download stats
        march_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-03-01:2023-03-31/{package_name}').json()
        april_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-04-01:2023-04-30/{package_name}').json()
        may_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-05-01:2023-05-31/{package_name}').json()
        june_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-06-01:2023-06-30/{package_name}').json()
        july_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-07-01:2023-07-31/{package_name}').json()
        aug_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-08-01:2023-08-31/{package_name}').json()
        sept_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-09-01:2023-09-30/{package_name}').json()
        oct_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-10-01:2023-10-31/{package_name}').json()
        nov_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-11-01:2023-11-30/{package_name}').json()
        dec_23 = requests.get(f'https://api.npmjs.org/downloads/range/2023-12-01:2023-12-31/{package_name}').json()
        jan_24 = requests.get(f'https://api.npmjs.org/downloads/range/2024-01-01:2024-01-31/{package_name}').json()
        feb_24 = requests.get(f'https://api.npmjs.org/downloads/range/2024-02-01:2024-02-29/{package_name}').json()
        
        # Extracting download stats
        march_23 = sum(day['downloads'] for day in march_23['downloads'])
        april_23 = sum(day['downloads'] for day in april_23['downloads'])
        may_23 = sum(day['downloads'] for day in may_23['downloads'])
        june_23 = sum(day['downloads'] for day in june_23['downloads'])
        july_23 = sum(day['downloads'] for day in july_23['downloads'])
        aug_23 = sum(day['downloads'] for day in aug_23['downloads'])
        sept_23 = sum(day['downloads'] for day in sept_23['downloads'])
        oct_23 = sum(day['downloads'] for day in oct_23['downloads'])
        nov_23 = sum(day['downloads'] for day in nov_23['downloads'])
        dec_23 = sum(day['downloads'] for day in dec_23['downloads'])
        jan_24 = sum(day['downloads'] for day in jan_24['downloads'])
        feb_24 = sum(day['downloads'] for day in feb_24['downloads'])
        
        data_row = {
            'package_name': package_name,
            'Mar 23': march_23,
            'Apr 23': april_23,
            'May 23': may_23,
            'Jun 23': june_23,
            'Jul 23': july_23,
            'Aug 23': aug_23,
            'Sep 23': sept_23,
            'Oct 23': oct_23,
            'Nov 23': nov_23,
            'Dec 23': dec_23,
            'Jan 24': jan_24,
            'Feb 24': feb_24
        }

        data_rows.append(data_row)
    
    except Exception as e:
        print(f"Error fetching data for {package_name}: {e}")

# Create a DataFrame from the list of rows
data = pd.DataFrame(data_rows)

# Save the DataFrame to a new CSV file
data.to_csv('download_trends_batch4.csv', index=False, escapechar='\\')
```
